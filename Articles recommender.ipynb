# Extracting the key words from the content description.

```python
#extracting key words
# initializing the new column
df3['Key_words'] = ""

for index, row in df3.iterrows():
    content = row['content']
    
    # instantiating Rake, by default is uses english stopwords from NLTK
    # and discard all puntuation characters
    r = Rake()

    # extracting the words by passing the text
    r.extract_keywords_from_text(content)

    # getting the dictionary whith key words and their scores
    key_words_dict_scores = r.get_word_degrees()
    
    # assigning the key words to the new column
    row['Key_words'] = list(key_words_dict_scores.keys())

# dropping the Plot column
df3.drop(columns = ['content'], inplace = True)
```

```python
#set title as the index 
df3.set_index('title', inplace = True)
df3.head()
```

```python
#creating bag of words
df3['bag_of_words'] = ''
columns = df3.columns
for index, row in df3.iterrows():
    words = ''
    for col in columns:
        if col != 'publication':
            words = words + ' '.join(row[col])+ ' '
        else:
            words = words + row[col]+ ' '
    row['bag_of_words'] = words
    
df3.drop(columns = [col for col in df3.columns if col!= 'bag_of_words'], inplace = True)
```

```python
# function that takes in article title as input and returns the top 3 recommended articles
def recommendations(title, cosine_sim = cosine_sim):
    
    recommended_articles = []
    
    # gettin the index of the article that matches the title
    idx = indices[indices == title].index[0]

    # creating a Series with the similarity scores in descending order
    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending = False)

    # getting the indexes of the 3 most similar articles
    top_3_indexes = list(score_series.iloc[1:4].index)
    
    # populating the list with the titles of the best 3 matching articles
    for i in top_3_indexes:
        recommended_articles.append(list(df3.index)[i])
        
    return recommended_articles
```
